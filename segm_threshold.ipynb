```python
import numpy as np
import matplotlib.pyplot as plt
```

**[Task 1.2 (a).]()** Use `plt.imread` to load an image:

```python
img = plt.imread("data/NIH3T3/im/dna-0.png") 
```

**[Task 1.2 (b).]()** Show the image:

```python
plt.figure("dna-0.png")
plt.imshow(img, "grey")
plt.colorbar() 
```

**[Task 1.2 (d).]()** Binarize the image by applying an intensity threshold:

```python
img_bin = img >= 0.10
plt.imshow(img_bin, interpolation= "None", cmap = "grey")
```

**[Task 1.2 (e).]()** Load the ground truth:

```python
img_ref = plt.imread("data/NIH3T3/gt/0.png")
dice = (2*np.sum(img_bin*img_ref))/(img_bin.sum()+img_ref.sum())
print(dice)
```
